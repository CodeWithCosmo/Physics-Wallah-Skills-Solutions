# 1
### Matplotlib: It is a library in python, which is used for data visualizaiton with the help of differnt charts and graphs present in the submodule pyplot of Matplotlib library.
#### Some of them are line plot, bar plot, histogram, pie chart, box plot etc...

### Example

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = np.arange(1,101,10)
```

```python
# line chart
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# bar chart
plt.bar(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# Scatter plot
x = np.arange(1,11)
y = np.random.randint(1,20,10)
plt.scatter(x,y,marker='*',s=100,c='purple')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# histogram
data = np.random.randint(1,11,20)
plt.hist(data,color='purple')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# Bos plot
data = np.random.randint(1,100,20)
plt.boxplot(data)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# 2
### Scatter plot: It is a type of graph used to display the relationship between two continous variable. It consists of a set of points, each representing a pair of values for the two variables being plotted.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.xlabel('X axis  ---->')
plt.ylabel('Y axis  ---->')
plt.show()
```

# 3 
### subplot() method is used to create multiple plots in a single output screen, This is used when we want to compare two or more data sets together.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])

y2 = np.array([50, 20, 40, 20, 60, 70])

y3 = np.array([10, 20, 30, 40, 50, 60])

y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
from matplotlib import pyplot as plt

plt.subplot(1,4,1)
plt.plot(x,y1,c = 'r')

plt.subplot(1,4,2)
plt.plot(x,y2,c='g')

plt.subplot(1,4,3)
plt.plot(x,y3,c='b')

plt.subplot(1,4,4)
plt.plot(x,y4,c='y')

plt.show()
```

# 4
### Bar plot: It is a type of data visualization that represents data using rectangualr bars. The length or height of each bar corresponds to the value it represets.

### They are used to compare the values of categorical data, summarizing information across different categories, statistics, business, finance etc..

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
from matplotlib import pyplot as plt
```

```python
# Ploting bar chart (vertcal)
plt.figure(figsize=(5,3))
plt.bar(company,profit,color = 'orange')
plt.show()
```

```python
# Ploting bar chart (horizontal)
plt.figure(figsize=(5,3))
plt.barh(company,profit,color ='purple')
plt.show()
```

### Box plot: It is also known as whisker plot, It is a graphical representation of a dataset that provides a summary of the distribution of the data.The plot consists of a box that shows the interquartile range(IQR) of the data, along with lines that extend from the box, called whiskers, that show the range of the data excluding outtliers.
### They are usefull for displaying the spread, skewwness, and outliners in a dataset, and for comparing the distribution of mutltiple datasets. They can be used to detect diffferences in the central tendency,variablity ,and shape of the distributions of two or more datasets. They commonly used in statistical analysis, quality control, and data exploration in a variety of fields.including madecine, finance,and social sciences.

```python
import numpy as np 
from matplotlib import pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
plt.show()
```
