```python
import plotly.graph_objects as go
import seaborn as sb
```

```python
df= sb.load_dataset('titanic')
df
```

```python
fig = go.Figure(go.Scatter(x=df.age,y=df.fare,mode='markers',marker=dict(size=5,color='purple')),
                go.Layout(title='Age vs Fare',xaxis=dict(title='Age'),yaxis=dict(title='Fare')))
fig.show()
```

## Displot:
### It allows us to create a histogram or density plot of a variable in our dataset. It provides a convenient way to visualize the distribution of a continous variable, and is useful for identifying the central tendency, variability and shape of the data.

```python
import plotly.express as px
```

```python
df = px.data.tips()
df
```

```python
fig = px.histogram(df,x='total_bill',nbins=50,opacity=0.4,marginal='rug',color_discrete_sequence=['purple'])
fig.update_layout(title='Distribution Plot',xaxis_title='Total Bill')
fig.show()
```
