### DataFrames
#### A DataFrame is a two-dimensional tabular data structure with labeled axes (rows and columns). It can be thought of as a spreadsheet or a SQL table. Each column in a DataFrame can have a different data type, such as integers, floats, or strings. A DataFrame can be created from a dictionary, a list of dictionaries, a NumPy array, or a CSV file.

### Series
#### A Series is a one-dimensional labeled array that can hold any data type (e.g. integers, strings, floats, etc.). A Series is essentially a single column of a DataFrame. It can be created from a dictionary, a list, a NumPy array, or a scalar value.

### Example

```python
print(type(dataFrame))
dataFrame
```

```python
ages = dataFrame['Age']
```

```python
print(type(ages))
ages
```

### Both Dataframes and Series are mutable, which means we can change them after theri creation.
### But Panel is immutable which can't be modified after creation.
### Example

```python
# DataFrames
record = pd.DataFrame({"Name":("Ram","Sita","Laxman"),"Age":(23,22,21)})
```

```python
print(type(record))
record
```

```python
record.at[3,('Name','Age')] = 'Bharat',19
```

```python
record # So, we have modified our original record 
```

```python
# Series
series = pd.Series([2,3,5,7,11,13,15])
```

```python
print(type(series))
series
```

```python
series.at[6] = 17
```

```python
series # We can modified the original series
```
